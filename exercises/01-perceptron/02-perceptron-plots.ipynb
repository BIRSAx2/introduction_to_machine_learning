```python
import matplotlib.pyplot as plt

dataset = [[1, 3.4, 1.0, -1], 
           [1, 1.5, 2.2, -1], 
           [1, 3.6, 2.4, 1], 
           [1, 1.7, 3.6, -1]]

w = [-1.0, -1.0, 1.0]

steps = 1000
```

```python
def plot(dataset, w):
  fig = plt.figure()
  plt.grid(linewidth=0.5)

  plt.plot(
    list(map(lambda x: x[1], filter(lambda x: x[3] == 1, dataset))),
    list(map(lambda x: x[2], filter(lambda x: x[3] == 1, dataset))),
    'v'
  )


  plt.plot(
    list(map(lambda x: x[1], filter(lambda x: x[3] == -1, dataset))),
    list(map(lambda x: x[2], filter(lambda x: x[3] == -1, dataset))),
    's'
  )

  plt.xlim(-5, 5)
  plt.ylim(-5, 5)

  x1 = -5 * (-w[1]/w[2]) - w[0]/w[2]
  x2 = +5 * (-w[1]/w[2]) - w[0]/w[2]
  plt.plot([-5, 5], [x1, x2], color='grey', linewidth = 1, linestyle = "--")
  
  
plot(dataset, w)
```

```python
num_errors = 0
steps = 1000
while steps > 0:
  for data_points in dataset:
    x = data_points[0:3]
    y = data_points[3]
    
    dot_product = w[0] * x[0] + w[1] * x[1] + w[2] * x[2]
    dot_product = sum([w[i] * x[i] for i in range(3)])
    
    if (dot_product >= 0 and y < 0) or (dot_product < 0 and y > 0):
      w[0] = w[0]  + (x[0] * y)
      w[1] = w[1]  + (x[1] * y)
      w[2] = w[2]  + (x[2] * y)
      num_errors += 1
  
  if num_errors == 0:
    break
  else: num_errors = 0
  steps = steps - 1
plot(dataset=dataset, w=w)
```
