# Polynomial Regression
This approach is based on the idea of using a polynomial function to model the relationship between the independent variable x and the dependent variable y. The polynomial function is given by:
$$
y = a_0 + a_1x + a_2x^2 + a_3x^3 + \ldots + a_nx^n
$$
where $a_0, a_1, a_2, \ldots, a_n$ are the coefficients of the polynomial function. The degree of the polynomial function is given by $n$.


In this notebook, we will use a second-degree polynomial function to model the relationship between the independent variable x and the dependent variable y. The second-degree polynomial function is given by:
$$
y = c + b + a^2
$$
where $a, b, c$ are the coefficients of the polynomial function.

### Implementation

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(314)
N = 250
x = np.random.exponential(scale = 1,size= N)


power = 2.5
noise = power * np.random.randn(N)

a = 4
b = -15
c = 1

y = a * x**2 + b * x + c + noise


# bias vector

x0 = np.ones(N)
x_new = np.vstack((x0, x, x**2)).T
plt.scatter(x, y, s = 2)
```

### Interpolation

The weights vector is computed in the same manner as in the linear regression. The weights vector is given by:
$$
W = (X^TX)^{-1}X^TY
$$
where $X$ is the matrix of the independent variable x, $Y$ is the vector of the dependent variable y, and $W$ is the weights vector.

```python
w = np.linalg.inv(x_new.T @ x_new) @ x_new.T @ y

print(f"Estimated parameters: a = {w[2]:.2f}, b = {w[1]:.2f}, c = {w[0]:.2f}")

x_frid = np.linspace(np.min(x), np.max(x), 100)
plt.scatter(x, y, s = 2)

plt.plot(x_frid, w[2] * x_frid**2 + w[1] * x_frid + w[0], 'r')
plt.show()
```
