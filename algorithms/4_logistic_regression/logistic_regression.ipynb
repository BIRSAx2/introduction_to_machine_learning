# Logistic Regression

Logistic regression is a statistical model that in its basic form uses a logistic function to model a binary dependent variable. In the regression analysis, the dependent variable is binary, i.e., it only has two possible outcomes. The goal of logistic regression is to find the best fitting model to describe the relationship between the dependent variable and a set of independent variables.

### Theoretical Aspects

In a binary classification problem, let $p$ be the probability of the positive class. The odds ratio is defined as the ratio of the probability of the positive class to the probability of the negative class. The odds ratio is given by:
$$
f(p) = \frac{p}{1-p}
$$

By plotting this function for $p \in [0, 1]$, we can see that the odds ratio ranges from 0 to $+\infty$. 

```python
import matplotlib.pyplot as plt
import numpy as np

p = np.arange(0.01, 0.99, 0.01)
odds_ratio = p / (1 - p)
plt.plot(p, odds_ratio)
plt.xlabel('p')
plt.ylabel('odds ratio')
plt.title('Odds ratio as a function of p')
plt.axvline(x=0.5, color='k', linestyle='--')
plt.axhline(y=1, color='k', linestyle='--')
```

### Logic function

We define the logit function as $f(p) = \log \left( \frac{p}{1-p} \right)$. The logit function maps the probability $p$ from the range $[0, 1]$ to the range $[-\infty, +\infty]$. 

```python
plt.plot(p, np.log(odds_ratio))

plt.axvline(x=0.5, color='k', linestyle='--')
plt.axhline(y=0, color='k', linestyle='--')
```

### Logistic *sigmoid* function

If $p$ is the probability that a point $x$ in the dataset is classified in the positive class $y = 1$, which can be written as $p(y = 1 | X_i)$ we can link this relationship with the logit function as follows:
$$
\ell\left[ p (y = 1 | X_i)\right] = w_0 X^0_i + w_1 X^1_i + \ldots + w_n X^n_i = w^T X_i = z
$$

$$

\begin{align*}
z &= \log\left( \frac{p}{1 - p} \right) \\

e^z &= \frac{p}{1 - p} \\

(1 - p) e^z &= p \\

e^z - pe^z &= p \\

e^z &= p(1 + e^z) \\

p &= \frac{e^z}{1 + e^z} \\

\end{align*}
$$


We can now define the sigmoid function as follows:
$$
\sigma(z) = \frac{1}{1 + e^{-z}}
$$

```python
def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))

z = np.arange(-7, 7, 0.1)
sigma_z = sigmoid(z)

# plot sigmoid
plt.plot(z, sigma_z)

# axis
plt.axvline(0.0, color='k', alpha = 0.5, linestyle = ":")
plt.axhline(0.5, color='k', alpha = 0.5, linestyle = ":")

# labels
plt.xlabel('z')
plt.ylabel('sigmoid(z)')

plt.axhline(1.0, color='g', alpha = 0.4, linestyle = "--")
plt.axhline(0.0, color='g', alpha = 0.4, linestyle = "--")
```
